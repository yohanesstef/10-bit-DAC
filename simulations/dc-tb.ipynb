## Python Script For Static Simulation Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_static_simulation(txt_file, corner):
    df = pd.read_csv(txt_file, sep=r'\s+')
    VOUT = df['load']
    n = df['ndata']
    plt.plot(n,VOUT)
    plt.title(f'Static {corner} Corner')
    plt.xlabel('n')
    plt.ylabel('VOUT(V)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
#def multi_plot():
```

```python
#txt_file = ['dc_tt-tb.txt', 'dc_ss-tb.txt', 'dc_ff-tb.txt', 'dc_fs-tb.txt', 'dc_sf-tb.txt']
txt_file = ['dc_tt_temp0-tb.txt', 'dc_tt_temp50-tb.txt', 'dc_tt_temp80-tb.txt', 'dc_tt_temp100-tb.txt']
#corner = ['TT','SS','FF','FS','SF']
corner = ['TT0','TT50','TT80','TT100']
for i in range(0,len(txt_file)):
    plot_static_simulation(txt_file[i], corner[i])
```
